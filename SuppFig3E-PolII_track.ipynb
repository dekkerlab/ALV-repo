```python
bw_fname = samples["polII"]["fname"]
chrom = "chr14"
start = 30_000_000
end = 40_000_000
binz = (-start+end)/25_000
track = bbi.fetch(
            bw_fname,
            chrom,
            start,
            end,
            bins=binz,
            missing=np.nan,
            summary="mean")

f,ax = plt.subplots(figsize=(8,2))
ax.plot(track,)
ax.set_xlim(0,binz)
ax.set_ylim(0,.6)

plt.savefig("SuppFig3E_PolII_chr14-30-40MB.pdf")
```
